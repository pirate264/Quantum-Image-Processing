```python
import numpy as np
import cv2
import time
import pandas as pd
import matplotlib.pyplot as plt

from skimage.metrics import mean_squared_error, peak_signal_noise_ratio, structural_similarity

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
```

```python
# Upload an image to Colab and set path
IMAGE_PATH = "/content/3047.jpg"

img = cv2.imread(IMAGE_PATH, cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (256, 256))
img = img / 255.0

plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")
```

```python
def frqi_edge(image):
    edge = np.zeros_like(image)
    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            edge[i,j] = abs(image[i,j] - image[i,j+1]) + abs(image[i,j] - image[i+1,j])
    return edge
```

```python
def neqr_edge(image, k=8):
    edge = np.zeros_like(image)
    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            p  = int(image[i,j]   * (2**k - 1))
            px = int(image[i,j+1] * (2**k - 1))
            py = int(image[i+1,j] * (2**k - 1))
            edge[i,j] = abs(p - px) + abs(p - py)
    return edge
```

```python
sim = AerSimulator()

def piqed_edge(image, shots=16):
    edge = np.zeros_like(image)

    # build + transpile ONCE
    qc = QuantumCircuit(1,1)
    qc.h(0)
    qc.rz(0.0, 0)
    qc.h(0)
    qc.measure(0,0)
    qc = transpile(qc, sim)

    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            delta = abs(image[i,j] - image[i,j+1]) + abs(image[i,j] - image[i+1,j])
            qc_local = qc.copy()
            qc_local.data[1][0].params = [2*np.pi*delta]

            result = sim.run(qc_local, shots=shots).result()
            counts = result.get_counts()
            edge[i,j] = counts.get('1',0) / shots

    return edge
```

```python
from qiskit.circuit import Parameter

def piqed_edge(image, shots=16):
    edge = np.zeros_like(image)

    # Define a parameter for the RZ gate
    phi = Parameter('phi')

    # build circuit ONCE with the parameter
    qc_param = QuantumCircuit(1,1)
    qc_param.h(0)
    qc_param.rz(phi, 0)   # Use parameter here
    qc_param.h(0)
    qc_param.measure(0,0)

    # Transpile the parameterized circuit once
    # Use optimization_level=0 to prevent changes that might break parameter binding if issues persist
    transpiled_qc = transpile(qc_param, sim, optimization_level=0)

    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            delta = abs(image[i,j] - image[i,j+1]) + abs(image[i,j] - image[i+1,j])

            # Bind the parameter to its value for the current pixel
            qc_bound = transpiled_qc.assign_parameters({phi: 2*np.pi*delta})

            result = sim.run(qc_bound, shots=shots).result()
            counts = result.get_counts()
            edge[i,j] = counts.get('1', 0) / shots

    return edge


frqi_img  = frqi_edge(img)
neqr_img  = neqr_edge(img)
piqed_img = piqed_edge(img)
```

```python
def evaluate(original, edge):
    mse  = mean_squared_error(original, edge)
    psnr = peak_signal_noise_ratio(original, edge, data_range=1)
    ssim = structural_similarity(original, edge, data_range=1)
    return mse, psnr, ssim
```

```python
metrics = {
    "FRQI":   evaluate(img, frqi_img),
    "NEQR":   evaluate(img, neqr_img),
    "PI-QED": evaluate(img, piqed_img)
}
```

```python
n = int(np.log2(256))
k = 8

def frqi_circuit(n):
    qc = QuantumCircuit(2*n + 1)
    for i in range(2*n):
        qc.h(i)
    return qc

def neqr_circuit(n, k):
    qc = QuantumCircuit(2*n + k)
    for i in range(2*n):
        qc.h(i)
    return qc

def piqed_circuit():
    qc = QuantumCircuit(1)
    qc.h(0)
    qc.rz(0.5, 0)
    qc.h(0)
    return qc
```

```python
circuits = {
    "FRQI": frqi_circuit(n),
    "NEQR": neqr_circuit(n, k),
    "PI-QED": piqed_circuit()
}

quantum_cost = {
    name: (qc.num_qubits, qc.depth())
    for name, qc in circuits.items()
}
```

```python
def encoding_time(build_func, *args):
    start = time.time()
    build_func(*args)
    return time.time() - start
```

```python
encoding_times = {
    "FRQI":   encoding_time(frqi_circuit, n),
    "NEQR":   encoding_time(neqr_circuit, n, k),
    "PI-QED": encoding_time(piqed_circuit)
}
```

```python
rows = []

for model in ["FRQI", "NEQR", "PI-QED"]:
    mse, psnr, ssim = metrics[model]
    qubits, depth   = quantum_cost[model]
    enc_time        = encoding_times[model]

    rows.append([
        model,
        qubits,
        depth,
        enc_time,
        mse,
        psnr,
        ssim
    ])

df = pd.DataFrame(rows, columns=[
    "Model",
    "Qubits Required",
    "Circuit Depth",
    "Encoding Time (s)",
    "MSE",
    "PSNR",
    "SSIM"
])

df
```
