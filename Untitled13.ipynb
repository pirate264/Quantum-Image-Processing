```python
def load_images(path, size=4, max_images=20):
    imgs, names = [], []
    # Ensure the directory exists. Create it if it doesn't.
    # Also create a dummy image to prevent an empty list error if no images are present.
    if not os.path.exists(path):
        os.makedirs(path)
        # Create a dummy image for testing
        dummy_image_path = os.path.join(path, "dummy_image.png")
        if not os.path.exists(dummy_image_path):
            dummy_image = np.zeros((size, size), dtype=np.uint8) # A black image
            cv2.imwrite(dummy_image_path, dummy_image)
        print(f"Created dummy dataset directory and image at {path}")


    files = [f for f in os.listdir(path)
             if f.lower().endswith(('.jpg','.png','.jpeg'))]

    print("Total image files found:", len(files))

    for f in files:
        img = cv2.imread(os.path.join(path, f), cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue
        img = cv2.resize(img, (size,size)) / 255.0
        imgs.append(img)
        names.append(f)
        if len(imgs) == max_images:
            break

    return np.array(imgs), names

X, names = load_images(DATASET_PATH, EVAL_SIZE, MAX_IMAGES)
print("Images loaded:", len(X))
```

```python
import numpy as np
from skimage.metrics import mean_squared_error, structural_similarity

EPS = 1e-10   # prevents infinite PSNR

def metrics(original, recon):
    original = np.nan_to_num(original)
    recon = np.nan_to_num(recon)

    # ---- MSE ----
    mse = mean_squared_error(original, recon)

    # ---- PSNR (manual, never inf) ----
    mse_safe = max(mse, EPS)
    psnr = 10 * np.log10((1.0 ** 2) / mse_safe)

    # ---- SSIM ----
    size = min(original.shape)
    win = size if size % 2 == 1 else size - 1
    win = max(3, win)

    if np.std(original) < EPS or np.std(recon) < EPS:
        ssim = 0.0
    else:
        ssim = structural_similarity(
            original,
            recon,
            data_range=1.0,
            win_size=win
        )

    return mse, psnr, ssim
```

```python
def gate_count(qc):
    return sum(qc.count_ops().values())
```

```python
from qiskit.quantum_info import Statevector

def safe_statevector(qc):
    state = Statevector.from_instruction(qc)
    data = np.asarray(state.data)

    if not np.all(np.isfinite(data)):
        raise RuntimeError("Invalid statevector")

    return state
```

```python
def safe_normalize_image(img):
    img = np.asarray(img, dtype=np.float64)
    img = np.nan_to_num(img)

    max_val = img.max()
    if max_val > EPS:
        return img / max_val
    return np.zeros_like(img)
```

```python
from qiskit import QuantumCircuit
import numpy as np

def nass_circuit(image):
    """
    NASS: Normalized Amplitude State Scheme
    SAFE against NaN / zero-norm errors
    """

    vec = image.flatten()
    amps = safe_amplitudes(vec)

    n = int(np.log2(len(amps)))
    qc = QuantumCircuit(n)

    qc.initialize(amps, qc.qubits)

    return qc
```

```python
def frqi_circuit(image):
    size = image.shape[0]
    n = int(np.log2(size))
    qc = QuantumCircuit(2*n + 1)

    qc.h(range(2*n))

    flat = image.flatten()
    for i, val in enumerate(flat):
        qc.cry(val * np.pi, i % (2*n), 2*n)

    return qc
```

```python
def efrqi_circuit(image):
    size = image.shape[0]
    n = int(np.log2(size))
    qc = QuantumCircuit(2*n + 1)

    qc.h(range(2*n))

    flat = image.flatten()
    for i, val in enumerate(flat):
        qc.cry(val * 2*np.pi, i % (2*n), 2*n)

    return qc
```

```python
def neqr_circuit(image, q=2):
    """
    NEQR with q-bit grayscale encoding
    SAFE against ambiguity errors
    """
    size = image.shape[0]
    n = int(np.log2(size))

    qc = QuantumCircuit(2*n + q)

    # Position superposition
    qc.h(range(2*n))

    # ✅ Force 1D integer pixel array
    flat = np.round(image.reshape(-1) * (2**q - 1)).astype(np.int64)

    for i in range(len(flat)):
        val = int(flat[i])  # ✅ force scalar
        for b in range(q):
            if ((val >> b) & 1):
                qc.cx(i % (2*n), 2*n + b)

    return qc
```

```python
def gqir_circuit(image, q=2):
    size = image.shape[0]
    n = int(np.log2(size))

    qc = QuantumCircuit(2*n + q)
    qc.h(range(2*n))

    # 🔑 FORCE 1D integer array
    flat = np.round(image.reshape(-1) * (2**q - 1)).astype(np.int64)

    for i in range(len(flat)):
        val = int(flat[i])   # 🔑 force scalar
        if val > 0:
            qc.cx(i % (2*n), 2*n)

    return qc
```

```python
def iqir_circuit(image):
    size = image.shape[0]
    n = int(np.log2(size))
    qc = QuantumCircuit(2*n + 1)

    qc.h(range(2*n))

    flat = image.flatten()
    for i, val in enumerate(flat):
        if val > 0.5:
            qc.cx(i % (2*n), 2*n)

    return qc
```

```python
def qrmw_circuit(image):
    size = image.shape[0]
    n = int(np.log2(size))
    qc = QuantumCircuit(2*n + 1)

    qc.h(range(2*n))

    flat = image.flatten()
    for i, val in enumerate(flat):
        qc.cry(val * np.pi, i % (2*n), 2*n)

    return qc
```

```python
def mcqi_circuit(image):
    size = image.shape[0]
    n = int(np.log2(size))
    qc = QuantumCircuit(2*n + 2)

    qc.h(range(2*n))

    flat = image.flatten()
    for i, val in enumerate(flat):
        qc.cry(val * np.pi, i % (2*n), 2*n)
        qc.cry((1-val) * np.pi, i % (2*n), 2*n + 1)

    return qc
```

```python
def qlr_circuit(image, threshold=0.5):
    size = image.shape[0]
    qc = QuantumCircuit(size * size)

    flat = image.flatten()
    for i, val in enumerate(flat):
        if val > threshold:
            qc.x(i)

    return qc
```

```python
def qsna_circuit(image):
    vec = image.flatten()
    amps = safe_amplitudes(vec)

    n = int(np.log2(len(amps)))
    qc = QuantumCircuit(n)
    qc.initialize(amps, qc.qubits)
    return qc
```

```python
phi = Parameter("phi")

def piqed_base_circuit():
    qc = QuantumCircuit(1,1)
    qc.h(0)
    qc.rz(phi,0)
    qc.h(0)
    qc.measure(0,0)
    return qc

qc_piqed = transpile(piqed_base_circuit(), backend, optimization_level=3)

def piqed_edge_fast(image, shots=16):
    edge = np.zeros_like(image)

    for i in range(1,image.shape[0]-1):
        for j in range(1,image.shape[1]-1):
            delta = abs(image[i,j]-image[i,j+1]) + abs(image[i,j]-image[i+1,j])
            bound = qc_piqed.assign_parameters({phi:2*np.pi*delta}, inplace=False)
            result = backend.run(bound, shots=shots).result()
            edge[i,j] = result.get_counts().get('1',0)/shots

    return edge
```

```python
models = {
    "FRQI": frqi_circuit,
    "EFRQI": efrqi_circuit,
    "NEQR": neqr_circuit,
    "GQIR": gqir_circuit,
    "IQIR": iqir_circuit,
    "QRMW": qrmw_circuit,
    "MCQI": mcqi_circuit,
    "NASS": nass_circuit,
    "QLR": None
}
```

```python
EPS = 1e-12

def safe_amplitudes(vec):
    vec = np.asarray(vec, dtype=np.float64)
    vec = np.nan_to_num(vec, nan=0.0, posinf=0.0, neginf=0.0)

    norm = np.linalg.norm(vec)

    if norm < EPS:
        out = np.zeros_like(vec)
        out[0] = 1.0        # |00...0> valid quantum state
        return out

    return vec / norm
```

```python
rows = []

piqed_info = piqed_base_circuit()

for img, name in zip(X, names):

    img = safe_normalize_image(img)

    for model, builder in models.items():

        start = time.perf_counter()
        qc = qlr_circuit(img) if model == "QLR" else builder(img)
        enc_time = time.perf_counter() - start

        state = safe_statevector(qc)

        recon = np.abs(state.data[:EVAL_SIZE * EVAL_SIZE])
        recon = safe_normalize_image(recon)
        recon = recon.reshape(EVAL_SIZE, EVAL_SIZE)

        mse, psnr, ssim = metrics(img, recon)

        rows.append({
            "Image": name,
            "Model": model,
            "Qubits": qc.num_qubits,
            "Depth": qc.depth(),
            "Gate Count": gate_count(qc),
            "Encoding Time": enc_time,
            "MSE": mse,
            "PSNR": psnr,
            "SSIM": ssim
        })

    # ---------- 2D-QSNA ----------
    start = time.perf_counter()
    qc = qsna_circuit(img)
    enc_time = time.perf_counter() - start

    state = safe_statevector(qc)
    recon = np.abs(state.data[:EVAL_SIZE * EVAL_SIZE])
    recon = safe_normalize_image(recon)
    recon = recon.reshape(EVAL_SIZE, EVAL_SIZE)

    mse, psnr, ssim = metrics(img, recon)

    rows.append({
        "Image": name,
        "Model": "2D-QSNA",
        "Qubits": qc.num_qubits,
        "Depth": qc.depth(),
        "Gate Count": gate_count(qc),
        "Encoding Time": enc_time,
        "MSE": mse,
        "PSNR": psnr,
        "SSIM": ssim
    })

    # ---------- PIQED ----------
    start = time.perf_counter()
    edge = piqed_edge_fast(img)
    enc_time = time.perf_counter() - start

    edge = safe_normalize_image(edge)
    mse, psnr, ssim = metrics(sobel(img), edge)

    rows.append({
        "Image": name,
        "Model": "PIQED (Proposed)",
        "Qubits": piqed_info.num_qubits,
        "Depth": piqed_info.depth(),
        "Gate Count": gate_count(piqed_info),
        "Encoding Time": enc_time,
        "MSE": mse,
        "PSNR": psnr,
        "SSIM": ssim
    })
```

```python
df = pd.DataFrame(rows)

summary = (
    df.groupby("Model")
      [["Qubits","Depth","Gate Count","Encoding Time","MSE","PSNR","SSIM"]]
      .mean()
      .round(6)
)

summary
```
